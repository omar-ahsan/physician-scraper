```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import lxml.html
import re
```

```python
doc_URLs = []

def getURLs(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    main_data = soup.find('div', class_="continuous-list-wrap")
    col = main_data.find_all('a', href=True)

    for c in col:
        doc_URLs.append('https://www.sirweb.org/' + c['href'])
    
    print(len(doc_URLs))
    print(doc_URLs)
    

```

```python
for i in range(1,18):
    link = f'https://www.sirweb.org/patient-center/doctor-finder/Results/?page={i}&Country=&State=&Expertise=Radiofrequency%20Ablation&Lastname=&Company=&class='
    getURLs(link)
```

```python
print(doc_URLs)

with open(r'C:\Users\Omar\Desktop\Web Scraping\doctors_up\rfa.txt', 'w') as fp:
    for doc_l in doc_URLs:
        # write each item on a new line
        fp.write("%s\n" % doc_l)
    print('Done')
```
