```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import lxml.html
import re
```

```python
Name_first = []

Name = []
Post = []
Contact = []
Location = []

flat_Name = []
flat_Contact = []
flat_Location = []
flat_Post = []
```

```python
def getData(link):
    N_F = []
    N = []
    P = []
    C = []
    L = []
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')
    information = soup.find('div', class_="continuous-list-wrap")

    doctor_col = information.find_all('p')
    for doc in doctor_col:
        name_element = doc.find('a')
        if name_element is not None:
            name = name_element.getText()
            N_F.append(name)

        phone = None
        phone_match = re.search(r'\(?\d{3}\)?[ .-]?\d{3}[ .-]?\d{4}', str(doc))
        if phone_match is not None:
            phone = phone_match.group()
            C.append(phone)
        else:
            C.append('-')

        city = None
        city_element = doc.find_previous('h4', class_='doctor__finder-city')
        if city_element is not None:
            city = city_element.getText()
            L.append(city.strip())

    for name in N_F:
        parts = name.split(',')
        if len(parts) > 1:
            N.append(parts[0].strip())
            P.append(', '.join(parts[1:]).strip())
        else:
            N.append(name.strip())
            P.append('')

    
    Name.append(N)
    Contact.append(C)
    Location.append(L)
    Post.append(P)


```

```python
for i in range(1, 18):
    url = f"https://www.sirweb.org/patient-center/doctor-finder/Results/?page={i}&Country=&State=&Expertise=Radiofrequency%20Ablation&Lastname=&Company=&class="
    getData(url)
```

```python
print(len(Name))
print(len(Post))
print(len(Contact))
print(len(Location))

for sublist in Name:
    for item in sublist:
        flat_Name.append(item)
for sublist in Post:
    for item in sublist:
        flat_Post.append(item)
for sublist in Contact:
    for item in sublist:
        flat_Contact.append(item)
for sublist in Location:
    for item in sublist:
        flat_Location.append(item)


print(len(flat_Name))
print(len(flat_Contact))
print(len(flat_Post))
print(len(flat_Location))
```

```python
pattern = r'\b([A-Z])[a-z]*\.\s?'

namess = [re.sub(pattern, '', name) for name in flat_Name]

print(namess)
print(len(namess))
```

```python
first_names = []
last_names = []

# Loop through each name in the list
for name in namess:
    # Split the name into a list of words
    words = name.split()
    # Extract the first word (i.e., the first name)
    first_name = words[0]
    # Extract the rest of the words (i.e., the last name)
    last_name = ' '.join(words[1:])
    # Append the first and last names to their respective lists
    first_names.append(first_name)
    last_names.append(last_name)

# Print the resulting lists
print("First names:", first_names)
print("Last names:", last_names)
```

```python
df = pd.DataFrame({'First Name': first_names,'Last Name': last_names, 'Practice Name': 'Radiofrequency ablation (RFA)', 'Qualifications': flat_Post, 'Contact': flat_Contact,
                   'City': flat_Location})

df.to_excel(r'physicians2.xlsx', index=False)
```
