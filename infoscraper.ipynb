```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
doc_urls = []
# open file and read the content in a list
with open(r'C:\Users\Omar\Desktop\Web Scraping\doctors_up\rfa.txt', 'r') as fp:
    for line in fp:
        # remove linebreak from a current name
        # linebreak is the last character of each line
        x = line[:-1]

        # add current item to the list
        doc_urls.append(x)


print(len(doc_urls))
print(doc_urls)
```

```python
hosp = []
street = []
zip_c = []

def getData(url):
    page = requests.get(url)
    time.sleep(2)
    soup = BeautifulSoup(page.content, 'html.parser')
    info = soup.find('div', class_="doctor-details")

    if info.find('strong', text='Practice or Primary Hospital:'):
        practice_or_hospital_strong = info.find('strong', text='Practice or Primary Hospital:')
        if practice_or_hospital_strong:
            # access the next sibling element, which is the <p> tag containing the address information
            address_p = practice_or_hospital_strong.find_next_sibling('p')
            # extract the text content of the address <p> tag
            address_text = address_p
            for br in address_p.find_all('br'):
                br.replace_with(',') # or br.replace_with(' ')
            address_text = address_p.getText().replace('\n', ' ').strip()

            clean_add_text = address_text.replace("\r", "").replace("  ", "")
            address_parts = clean_add_text.split(',')
            print('Address part : ', len(address_parts))
            hosp.append(address_parts[0])

            if len(address_parts) == 4:
                street.append(address_parts[1])
            elif len(address_parts) == 5:
                street.append(address_parts[1] + ',' + address_parts[2])
            elif len(address_parts) == 6:
                street.append(address_parts[1] + ',' + address_parts[2] + ',' + address_parts[3])
            elif len(address_parts) == 7:
                street.append(address_parts[1] + ',' + address_parts[2] + ',' + address_parts[3] + ',' + address_parts[4])
            elif len(address_parts) == 8:
                street.append(address_parts[1] + ',' + address_parts[2] + ',' + address_parts[3] + ',' + address_parts[4] + ',' + address_parts[5])
            
            zip_code = address_parts[len(address_parts) - 1].strip().split(' ')[-1]
            zip_c.append(zip_code)
    else:
        print('Practice or Primary Hospital: not found')
        hosp.append('-')
        street.append('-')
        zip_c.append('-')
```

```python
i = 0
for link in doc_urls:
    getData(link)
    print(i)
    i = i + 1
    

```

```python

print(len(hosp))
print(len(street))
print(len(zip_c))

hosp.pop()
zip_c.pop()

df = pd.DataFrame({'Hospital or Primary Hospital': hosp, 'Street or Office Address': street, 'Zip Code': zip_c})
df.to_excel(r'address_test2.xlsx', index=False)
```
